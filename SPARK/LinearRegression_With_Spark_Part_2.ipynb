```
from pyspark.ml.regression import LinearRegression
```

```
training = spark.read.format("libsvm").load("sample_linear_regression_data.txt")
```

```
training.show()
```

```
lreg = LinearRegression(featuresCol='features', labelCol='label', predictionCol='predictions')
lrModel = lreg.fit(training)
```

```
print ("The coefficients - ", lrModel.coefficients)
print ("The intercepts - ", lrModel.intercept)
```

```
# Get the summary
training_summary = lrModel.summary

print ("R2 score -", training_summary.r2)
print ("RMSE score -", training_summary.rootMeanSquaredError)
```

```
all_data = spark.read.format("libsvm").load("sample_linear_regression_data.txt")
train_data, test_data = all_data.randomSplit([0.7,0.3])
print (train_data.describe().show())
print (test_data.describe().show())
```

Following are the observations from the above code block - 
*   We have divided the whole dataset into training and validation datasets.
*   The training data consists of 349 records.
*   The testing data consists of 152 records.

```
correctModel = lreg.fit(train_data)
testResults = correctModel.evaluate(test_data)
```

```
print ("R2 score -", testResults.r2)
print ("RMSE score -", testResults.rootMeanSquaredError)
print ("MAE score -", testResults.meanAbsoluteError)
```

```
predictions = lrModel.transform(unlabeled_data)
predictions.show()
```

```
cust_df = spark.read.csv("Ecommerce_Customers.csv", inferSchema=True, header=True)
cust_df.show()
```

```
cust_df.printSchema()
```

```
# Target variable = Yearly Amount Spent 
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors

cust_df.columns
```

```
assembler = VectorAssembler(inputCols=['Avg Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent'],
                            outputCol='features')

output = assembler.transform(cust_df)
output.head(5)
```

```
finalData = output.select('features','Yearly Amount Spent')
finalData.show()
```

```
train_data, test_data = finalData.randomSplit([0.7,0.3])
print (train_data.describe().show())
print (test_data.describe().show())
```

```
lreg = LinearRegression(featuresCol='features', labelCol='Yearly Amount Spent', predictionCol='predictions')
lrModel = lreg.fit(train_data)
```

```
testResults = lrModel.evaluate(test_data)
```

```
print ("R2 score -", testResults.r2)
print ("RMSE score -", testResults.rootMeanSquaredError)
print ("MAE score -", testResults.meanAbsoluteError)
```

```
ship_df = spark.read.csv('cruise_ship_info.csv', inferSchema=True, header=True)
ship_df.show(5)
```

```
ship_df.columns
```

```
ship_df.groupBy('Cruise_line').count().show()
```

```
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol='Cruise_line', outputCol = 'cruise_category')
indexed = indexer.fit(ship_df).transform(ship_df)
indexed.show(5)
```

```
assembler = VectorAssembler(inputCols=['Age', 'Tonnage', 'passengers', 'length', 'cabins', 'passenger_density', 'cruise_category'],
                            outputCol='features')

output = assembler.transform(indexed)
output.show(5)
```

```
finalData = output.select('features', 'crew')
finalData.show()
```

```
train_data, test_data = finalData.randomSplit([0.7,0.3])
print (train_data.describe().show())
print (test_data.describe().show())

lreg = LinearRegression(featuresCol='features', labelCol='crew', predictionCol='predictions')
lrModel = lreg.fit(train_data)

testResults = lrModel.evaluate(test_data)

print ("R2 score -", testResults.r2)
print ("RMSE score -", testResults.rootMeanSquaredError)
print ("MAE score -", testResults.meanAbsoluteError)
```
